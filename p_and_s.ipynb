```python
import numpy as np
import matplotlib.pyplot as plt

mu, sigma = 100, 5
data = np.random.normal(mu, sigma, 1000)

plt.hist(data, bins=30, edgecolor='black', alpha=0.7)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Normally Distributed Data')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mu, sigma = 100, 5
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
pdf = norm.pdf(x, mu, sigma)

plt.plot(x, pdf, label="Normal Distribution", color="blue")

plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.title("Probability Density Function (PDF) of a Normal Distribution")
plt.axvline(mu, color='red', linestyle='dashed', label="Mean (μ)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
from scipy.stats import norm # Import norm from scipy.stats
# Define the points x1 and x2
x1 = -1
x2 = 1
# Generate the x values
x = np.linspace(-4, 4, 1000)
# Generate the y values for the normal distribution
y = norm.pdf(x) # Use norm directly instead of stats.norm
# Create the plot
plt.plot(x, y, label='Normal Distribution')
# Fill the area for P(x1 < X < x2)
plt.fill_between(x, y, where=(x > x1) & (x < x2), color='gray', alpha=0.5, label='P(x1 < X < x2)')
# Fill the area for P(X < x2)
plt.fill_between(x, y, where=(x < x2), color='blue', alpha=0.3, label='P(X < x2)')
# Fill the area for P(x1 < X)
plt.fill_between (x, y, where=(x > x1), color='green', alpha=0.3, label='P(x1 < X)')
# Add labels and title
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('Normal Distribution with Shaded Areas')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import scipy.stats as stats

# Define parameters
mu = 100  # Mean IQ
sigma = 15  # Standard deviation

# Take user inputs for IQ values
iq1 = float(input("Enter first IQ value (IQ1): "))
iq2 = float(input("Enter second IQ value (IQ2): "))

# Compute probabilities
prob_iq1 = stats.norm.cdf(iq1, mu, sigma)
prob_iq2 = stats.norm.cdf(iq2, mu, sigma)

# Compute probability between IQ1 and IQ2
prob_between = prob_iq2 - prob_iq1

# Print results
print(f"P(IQ ≤ {iq1}) = {prob_iq1:.4f}")
print(f"P(IQ ≤ {iq2}) = {prob_iq2:.4f}")
print(f"P({iq1} ≤ IQ ≤ {iq2}) = {prob_between:.4f}")
```

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Define parameters
mu = 100  # Mean IQ
sigma = 15  # Standard deviation

# Define IQ values
x1 = 85
x2 = 115

# Calculate probabilities
p_less_85 = stats.norm.cdf(x1, mu, sigma)  # P(IQ ≤ 85)
p_greater_85 = 1 - p_less_85  # P(IQ ≥ 85)
p_between_85_115 = stats.norm.cdf(x2, mu, sigma) - p_less_85  # P(85 ≤ IQ ≤ 115)

# Print probabilities
print(f"P(IQ ≥ {x1}) = {p_greater_85:.4f}")
print(f"P(IQ ≤ {x1}) = {p_less_85:.4f}")
print(f"P({x1} ≤ IQ ≤ {x2}) = {p_between_85_115:.4f}")

# Visualization
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
y = stats.norm.pdf(x, mu, sigma)

plt.figure(figsize=(8, 5))
plt.plot(x, y, label="Normal Distribution", color="blue")

# Shade areas under the curve
plt.fill_between(x, y, where=(x >= x1), color="green", alpha=0.3, label=f"P(IQ ≥ {x1})")
plt.fill_between(x, y, where=(x <= x1), color="red", alpha=0.3, label=f"P(IQ ≤ {x1})")
plt.fill_between(x, y, where=((x >= x1) & (x <= x2)), color="purple", alpha=0.5, label=f"P({x1} ≤ IQ ≤ {x2})")

# Labels and legend
plt.xlabel("IQ Score")
plt.ylabel("Probability Density")
plt.title("IQ Distribution and Probability Regions")
plt.legend()
plt.show()
```
